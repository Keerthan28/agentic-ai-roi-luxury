```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', 50)
sns.set_theme(style='whitegrid')
```

## 1. Load the model results

`results/roi_model_results.csv` is produced by running `python src/agentic_ai_roi_model.py` end-to-end. Every row is one customer with the engineered features and predicted ROI components.

```python
df = pd.read_csv('../results/roi_model_results.csv')
print(df.shape)
df.head()
```

## 2. Headline ROI table

```python
df[['net_roi', 'ai_readiness', 'retention_improvement', 'clv',
    'ai_revenue_uplift', 'retention_savings', 'ai_cost']].describe()
```

## 3. Distribution of predicted ROI

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.hist(df['net_roi'], bins=25, color='#2c3e50', edgecolor='white', alpha=0.85)
ax.axvline(df['net_roi'].mean(), color='#e74c3c', linestyle='--', label=f"Mean: {df['net_roi'].mean():.2f}x")
ax.set_xlabel('Net ROI (x)')
ax.set_ylabel('Frequency')
ax.set_title('Predicted Net ROI per Customer')
ax.legend()
plt.tight_layout()
```

## 4. What-if: halve the personalization revenue lift

The 20% personalization lift is an upper-bound industry assumption. What happens to ROI if we halve it?

```python
half_uplift = df['ai_revenue_uplift'] / 2
stressed_roi = (half_uplift + df['retention_savings'] - df['ai_cost']) / df['ai_cost']
print(f"Base mean ROI:         {df['net_roi'].mean():.2f}x")
print(f"Stressed mean ROI:     {stressed_roi.mean():.2f}x")
print(f"Share still positive:  {(stressed_roi > 0).mean():.0%}")
```
